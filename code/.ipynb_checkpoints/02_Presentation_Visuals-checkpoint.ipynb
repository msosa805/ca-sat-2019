# Visuals

## Percent Diversity vs. Percent of Test Takers

```python
plt.figure(figsize = (12, 5))
plt.scatter(df['PctDiv'],
            df['PctTstTakers'],
            color = '#D55E00')

plt.title('Percent of students who took the SAT in each district by percent diversity')

plt.xlabel('% non-white or latino')
plt.ylabel('% of students who took SAT')

plt.savefig('images/percent_div_vs_test_takers.png');
```

## Percent Diversity vs. Percent Math Benchmark

```python
plt.figure(figsize = (12, 5))
plt.scatter(df['PctDiv'],
            df['PctMathBenchmark12'], color = '#FFC20A')

plt.title('Percent of students in each district who passed MATH benchmark, by percent diversity')

plt.xlabel('% non-white or latino')
plt.ylabel('% passed math benchmark')

# SOURCE: https://www.statology.org/scatterplot-with-regression-line-python/
# m (slope) and b (intercept) of linear regression line
m, b = np.polyfit(df['PctDiv'],
                  df['PctMathBenchmark12'],1)
# add linear regression line to scatterplot
x = df['PctDiv']
plt.plot(x, m*x+b, color = '#a7bbc7')

plt.savefig('images/percent_div_vs_math.png')
```

## Percent Diversity vs. Percent English Benchmark

```python
plt.figure(figsize=(12, 5))
plt.scatter(df['PctDiv'],
            df['PctERWBenchmark12'], color = '#0C7BDC')

plt.title('Percent of students in each district who passed ENGLISH benchmark, by percent diversity')

plt.xlabel('% non-white or latino')
plt.ylabel('% passed reading benchmark')

m, b = np.polyfit(df['PctDiv'],
                  df['PctERWBenchmark12'], 1)

plt.plot(x, m*x+b, color = '#a7bbc7')

plt.savefig('images/percent_div_vs_english.png');
```

## Percent Diversity vs. Percent of Both Benchmark Tests

```python
plt.figure(figsize = (12, 5))
plt.scatter(df['PctDiv'],
            df['PctBothBenchmark12'], color = '#56B4E9')

plt.title('Percent of students in each district who passed BOTH math and reading benchmark, by percent diversity')

plt.xlabel('% non-white or latino')
plt.ylabel('% of students who pass BOTH tests')

m, b = np.polyfit(df['PctDiv'],
                  df['PctBothBenchmark12'], 1)

plt.plot(x, m*x+b, color = '#a7bbc7')

plt.savefig('images/percent_div_vs_both.png');
```
